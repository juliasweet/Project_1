```python
file1 = "Economic_vs_Military.xlsx"
file2 = "GDP_by_$.xlsx"

GDP_df = pd.read_excel(file2)
Aid_df = pd.read_excel(file1)
Aid_df.head()
```

```python
GDP_df = GDP_df.rename(columns={"Total":"GDP","Country Name":"Country","Year":"Year"})
GDP_df.head()
```

```python
GDP_plot_df=GDP_df.set_index('Year', drop=True)
GDP_plot_df
```

```python
Combined_data = pd.merge(Aid_df,GDP_plot_df, on=['Country','Year'])
Combined_data
```

```python
Combined_Aid_GDP = Combined_data.rename(index=str, columns={"Grand Total": "Total Aid"})

Combined_Aid_GDP.head(10)
```

```python
from sklearn import linear_model
X0 = Combined_Aid_GDP[['Economic','Military']] # here we have 2 variables for multiple regression. If you just want to use one variable for simple linear regression, then use X = df['Economiv'] for example. Alternatively, you may add additional variables within the brackets
Y0 = Combined_Aid_GDP['GDP']

# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X0, Y0)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```

```python
import statsmodels.api as sm
# with statsmodels
X = sm.add_constant(X0) # adding a constant 
model = sm.OLS(Y0, X).fit()
predictions = model.predict(X) 
 
print_model = model.summary()
print(print_model)
```
