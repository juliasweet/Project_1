```python
file1 = "Economic_vs_Military.xlsx"
file2 = "StabilityScore.csv"

Stability_df = pd.read_csv(file2)
Aid_df = pd.read_excel(file1)
Aid_df.head()
```

```python
Aid_df = Aid_df.rename(columns={"Row Labels": "Country"})
Aid_df.head()
```

```python
Stability_df.head()
```

```python
Stability_df = Stability_df.rename(columns={"AO":"Angola", "BI": "Burundi", "BJ": "Benin", "ET":"Ethiopia", "GH": "Ghana", "KE":"Kenya", "MG": "Madagascar", "ML":"Malawi", "MZ":"Mozambique",  "NA": "Namibia", "NG":"Nigeria", "RW":"Rwanda", "SN": "Senegal", "SL":"Sierra Leone", "SO": "Somalia", "UG":"Uganda", "ZA":"South Africa", "ZW":"Zimbabwe"})
Stability_df.head()
```

```python
Stability_df = Stability_df.rename(columns={"Unnamed: 0": "Year"})
Stability_df.head()
```

```python
Stability_df.set_index('Year', drop=True)
```

```python
#pivot columns into rows
Stability_pivot = Stability_df.melt(id_vars=["Year"], 
        var_name="Country", 
        value_name="Total Stability")
Stability_pivot.head(10)
```

```python
Combined_Aid_Stability = pd.merge(Aid_df, Stability_pivot, on=['Country','Year'])
Combined_Aid_Stability.set_index('Country', drop=True)

Combined_Aid_Stability.head(10)
```

```python
#x_axis = Year
#y = stability
#x1 = Total Aid # size of the bubble
```

```python
#pip install cufflinks in GitBash

Combined_Aid_Stability.iplot(kind='bubble', x='Year', y='Total Stability', size='Grand Total', text='Country',
             xTitle='Year', yTitle='Stability Index',
             filename='Aid_vs_Stability_bubbleplot')
```
