```python
import wbpy
from pprint import pprint

api = wbpy.IndicatorAPI()

iso_country_codes = ["KEN", "ZAF", "NGA", "ZWE", "ETH", "UGA",
                   "GHA", "SEN", "MDG", "MOZ", "NAM", "BDI",
                  "BEN", "SOM", "MLI", "AGO", "SLE", "RWA"]
total_population = "SP.POP.TOTL"

dataset_popl = api.get_dataset(total_population, iso_country_codes, date="2001:2017")
dataset_popl
```

```python
d6=dataset_popl.as_dict()
```

```python
Populationtotal6_df=pd.DataFrame(data=d6)
```
