```python
file1 = "Economic_vs_Military.xlsx"
file2 = "GDP_by_$.xlsx"

GDP_df = pd.read_excel(file2)
Aid_df = pd.read_excel(file1)
Aid_df.head()
```

```python
Aid_df = Aid_df.rename(columns={"Row Labels": "Country"})
Aid_df.head(10)
```

```python
Aid_df['Country'].value_counts()
```

```python
GDP_df.head()
```

```python
GDP_df = GDP_df.rename(columns={"Total":"GDP","Country Name":"Country","Year":"Year"})
GDP_df.head()
```

```python
GDP_plot_df=GDP_df.set_index('Year', drop=True)
GDP_plot_df
```

```python
Combined_data = pd.merge(Aid_df,GDP_plot_df, on=['Country','Year'])
Combined_data
```

```python
Combined_Aid_GDP = Combined_data.rename(index=str, columns={"Grand Total": "Total Aid"})

Combined_Aid_GDP
```

```python
Combined_Aid_GDP.iplot(kind='bubble', x='Year', mode='markers', y='GDP', size='Total Aid', text='Country',
             xTitle='Year', yTitle='GDP',
             filename='Aid_vs_GDP_bubbleplot')
```

```python
Combined_Aid_GDP.iplot(kind='line', x='Year', mode='markers', y='GDP', text='Country',
             xTitle='Year', yTitle='GDP',
             filename='Aid_vs_GDP_bubbleplot')
```
