To get stability data, I employed wbpy, a wrapper for the World Bank API. This allowed me to easily pull data from the World Bank API. With the code in the second cell, we could easily get additional data on other indicators by changing the indicator for which we wanted data. The code could also be easily adapted to search other countries. 

We included only countries that were classed as being part of the sub-Saharan African region in both the World Bank API and the USAID data. 

```python
#Dependencies
import pandas as pd
import numpy as np
import wbpy
from pprint import pprint
```

```python
api = wbpy.IndicatorAPI()

iso_country_codes = ["KEN", "ZAF", "NGA", "ZWE", "ETH", "UGA",
                     "GHA", "SEN", "MDG", "MOZ", "NAM", "BDI",
                    "BEN", "SOM", "MLI", "AGO", "SLE", "RWA"]
stability = "PV.EST"

dataset_EST = api.get_dataset(stability, iso_country_codes, date="2002:2017")
dataset_EST
```

## After importing the data, I converted it to a dictionary, which I then turned into a DataFrame. 

```python
d=dataset_EST.as_dict() 
```

```python
stabilityscore_df=pd.DataFrame(data=d)
```
